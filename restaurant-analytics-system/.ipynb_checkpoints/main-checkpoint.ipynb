```python
import pandas as pd
import numpy as np
from rich import print
from rich.console import Console
from rich.table import Table
```

```python
load_csv = pd.read_csv(r"C:\Users\shobh\Downloads\restaurant_analysis app\real_indian_restaurants_dataset.csv")

df = pd.DataFrame(load_csv)
```

```python
def printrestaurant():
    
    table = Table(title="🍽 All Restaurants")
    
    for column in load_csv.columns:
        table.add_column(column)
    
    for _, row in load_csv.iterrows():
        table.add_row(*[str(value) for value in row])
    console.print(table)

```

```python
input("Press any key to continue :")
while True:
    try:
            
        print('''
    ========================================
     🍴 RESTAURANT ANALYTICS SYSTEM
    ========================================
    
    Welcome User!
    
    1. View All Restaurants
    2. Search Restaurants by City
    3. Filter by Rating
    4. Filter by Budget
    5. Show Top Rated Restaurants
    6. Show Cheapest Restaurants
    7. City Analytics
    8. Exit''')
    
    
        choose=int(input("Enter your choice:"))
        if choose==1:
            print("\nPrinting all resturants:\n")
            printrestaurant()
        elif choose==2:
            city = input("Enter your city:")
    
            result = df[df["City"]==city]    [["Restaurant","City"]]
            
            result.index =range(1,len(result)+1)
            
            if len(result)>0:
                print("\n========================================\n")
                print(F"{len(result)} Matches Found")
                print(result)
            else:
                print("\n========================================\n")
                print("Sorry! No Resturant Found in city",city)
                print("\n========================================\n")
        elif choose==3:
            print("\n========================================\n")
            print("\nAll Restaurant Filtered By Rating\n")
            fltr = df.sort_values(by="Rating", ascending=True)
    
            fltr.index = range(1, len(fltr) +1)
            
            print(fltr)
            print("\n========================================\n")
        elif choose==4:
            print("""
    1. Filter Highest To Lowest
    2. Filter Lowest To Highest""")
            ch=int(input("Enter 1 or 2:"))
            if ch==1:
                print("\n========================================\n")
                print("\n Filtering Restaurant On The Basis of Budget\n")
                print("[red] Highest[/red] to [green]Lowest[/green]")
                bud = df.sort_values(by="Price",ascending =False)
    
                bud.index=range(1,len(bud)+1)
                
                print(bud)
            if ch==2:
                print("\n========================================\n")
                print("\n Filtering Restaurant On The Basis of Budget\n")
                print("[green] Lowest[/green] to [bold red]Highest[/bold red]")
                bud = df.sort_values(by="Price",ascending =True)
                bud.index=range(1,len(bud)+1)
                print(bud)
            else:
                print("Invalid Input Recived ! Please Enter only 1 or 2")
        elif choose==5:
            print("""
    1. Check For Any Specific City
    2. Check Across Country""")
            choo = int(input("Enter your choice  (1/2):"))
            if choo ==1:
                ent = input("Enter your city:")
                
                res = df[df["City"]==ent] [["Restaurant","City","Rating"]]
                
                res.index = range(1,len(res)+1)
                
                rated = res.sort_values(by="Rating",ascending=False)
                
                
                rated.index = range(1,len(rated)+1)
                
                print(rated.head(5))
            elif choo==2:
                rated = df.sort_values(by="Rating",ascending=False)
    
                rated.index = range(1,len(rated)+1)
                
                print(rated.head(5))
            else:
                print("Invalid Operation . Please Enter Only 1 or 2")
        elif choose==6:
            print("\n========================================\n")
            print("Cheapest Restaurants⬇️")
            bud = df.sort_values(by="Price",ascending =True)
            
            bud.index=range(1,len(bud)+1)
            
            print(bud.head(5))
            print("\n========================================\n")
        elif choose==7:
            print("\n========================================\n")
            print("City Analytics")
            
            #Average resturnt ratingas 
    
    
            rated = df.groupby("City")["Rating"].mean().round(1)
            
            print("\n Average Restaurant Rating :\n")
            print(rated.to_string())
            
            print("\n========================================\n")
            
            
            sort = df.sort_values(by="Price",ascending=False)
            
            sort.index  = range(1, len(sort) +1)
            
            print("Most Expensive Resturant :\n\n ",sort.head(1))
            
            print("\n========================================\n")
            
            sort = df.sort_values(by="Price",ascending=True)
            
            sort.index  = range(1, len(sort) +1)
            
            print("\nMost Cheapest Resturant :\n\n ",sort.head(1))
    
            print("\n========================================\n")
    
    
        elif choose==8:
            print("""
    Thank you for using Restaurant Analytics System!
    
    Have a great meal """)
            break
        
        else:
            print("Invalid Operation Performed ! Please Select from 1 -8 ")

    except:
        print("Please Enter From 1 to 8 NUMBERS")



        
    
    
    
            
            
            
            
                            
                        
    
            
```
